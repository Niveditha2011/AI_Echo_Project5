```python
import pandas as pd
import re
import string
from langdetect import detect
import spacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Define basic English stopwords manually
basic_stopwords = {
    'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your',
    'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', 'her',
    'hers', 'herself', 'it', 'its', 'itself', 'they', 'them', 'their', 'theirs',
    'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', 'these', 'those',
    'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had',
    'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if',
    'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with',
    'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after',
    'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over',
    'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where',
    'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other',
    'some', 'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so', 'than', 'too',
    'very', 'can', 'will', 'just', 'don', 'should', 'now'
}

# Preprocessing function
def preprocess(text):
    if pd.isna(text):
        return None

    # Language detection (optional)
    try:
        if detect(text) != 'en':
            return None
    except:
        return None

    # Lowercase
    text = text.lower()

    # Remove special characters, digits, and punctuation
    text = re.sub(r'[^a-z\s]', '', text)

    # Tokenize and lemmatize using spaCy
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if token.lemma_ not in basic_stopwords and token.is_alpha]

    # Join back to a single string
    return ' '.join(tokens)

# Load your dataset
df = pd.read_csv("your_file.csv")

# Handle missing reviews
df = df[df['review'].notna()].copy()

# Apply preprocessing
df['cleaned_review'] = df['review'].apply(preprocess)

# Drop rows where preprocessing returned None (non-English or empty)
df = df[df['cleaned_review'].notna()].copy()
```
